```python
#standardization
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
x_train
```

```python
#z-score
from scipy.stats import zscore
zscore(x)
```

```python
#K neigbhor
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train,y_train)
y_pred=knn.predict(x_test)
y_pred
x_test[3]
```

```python
# Setup arrays to store training and test accuracies
neighbors = np.arange(1, 9)
train_accuracy = np.empty(len(neighbors))
test_accuracy = np.empty(len(neighbors))
test_accuracy
```

```python
train_accuracy
```

```python
for i,k in enumerate(neighbors):
  knn=KNeighborsClassifier(n_neighbors=k)
  knn.fit(x_train,y_train)
  train_accuracy[i]=knn.score(x_train,y_train)
  test_accuracy[i]=knn.score(x_test,y_test)
  print('Neighbors: %d, Training Accuracy: %f, Testing Accuracy: %f' % (k,train_accuracy[i],test_accuracy[i]))
```

```python
import matplotlib.pyplot as plt
# Generate plot
plt.figure(figsize=(10,6)) # Use plt instead of plot
plt.plot(neighbors, test_accuracy, label = 'Testing Accuracy') # Use plt instead of plot
plt.plot(neighbors, train_accuracy, label = 'Training Accuracy') # Use plt instead of plot
plt.legend()
plt.title('k-NN Varying number of neighbors')
plt.xlabel('Number of neighbors')
plt.ylabel('Accuracy')
plt.show()
```
